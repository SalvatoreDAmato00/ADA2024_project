```python
import pandas as pd
import os

# Identify Files
file_directory = r'C:\Users\sdama\Downloads\TEMPERATURES'
file_names = [
    'KANSAS_1990_2020.xlsx', 'MISSOURI_1990_2020.xlsx', 'WISCONSIN_1990_2020.xlsx', 
    'OHIO_1990_2020.xlsx', 'SOUTH_DAKOTA_1990_2020.xlsx', 'INDIANA_1990_2020.xlsx', 
    'NEBRASKA_1990_2020.xlsx', 'MINNESOTA_1990_2020.xlsx', 'ILLINOIS_1990_2020.xlsx', 
    'IOWA_1990_2020.xlsx'
]  # List of names of files

# Read Files and Assign Region
dfs = []
for file_name in file_names:
    file_path = os.path.join(file_directory, file_name)
    df = pd.read_excel(file_path)
    # Extract state name from file_name by splitting and removing non-state text
    state_name = file_name.split('_')[0]
    # Assign the state name to a new column in the DataFrame
    df['Region'] = state_name
    dfs.append(df)

# Merge DataFrames
temperature_df = pd.concat(dfs, ignore_index=True)

# Save Merged Data to File
temperature_df.to_csv('cornbelt_data2.csv', index=False)

# Display the merged DataFrame
temperature_df
```
